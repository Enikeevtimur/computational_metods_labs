```python
import numpy as np
import math 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def fun(x):
    return math.cos(x)/(x+2)
```

```python
a = 0.4
b = 1.2
eps = 10**(-6)
```

```python
def d_fun_e(x,h):
    return (fun_e(x+h)-fun_e(x))/h
```

Обобщенная формула левых прямоугольников


```python
df = pd.DataFrame({'length':[], 'epsilons': [], 'p value': [], 'integral value': []})
df1 = pd.DataFrame({'length':[], 'epsilons': [], 'p value': [], 'integral value': []})
df2 = pd.DataFrame({'length':[], 'epsilons': [], 'p value': [], 'integral value': []})
```

```python
def new_x(h,n):
    arr = []
    for i in range(n+1):
        arr.append(a+i*h)
    return arr
```

```python
def pr(arr,n):
    integral = 0
    for i in range(1,n+1):
        integral += (arr[i]-arr[i-1])*fun(arr[i-1])
    return integral
```

```python
cur = pr(new_x((b-a)/1,1),1)
prev = 0
i = 1
while abs(cur-prev) > eps:
    i *= 2
    e = abs(cur - prev)
    prev = cur 
    cur = pr(new_x((b-a)/i,i),i)
    if i == 2:
        df2.loc[len(df2.index)] = [i,abs(cur-prev),0,cur]
    else:
        df2.loc[len(df2.index)] = [i,abs(cur-prev),-math.log2(abs(cur-prev)/e),cur]
    print(cur,prev,i)
df2
```

Теперь возьмем формулу Симпсона

```python
def new_x_s(n,h):
    arr = []
    for i in range(2*n+1):
        arr.append(a+i*h)
    return arr
```

```python
new_x_s(2,(b-a)/2*2)
```

```python
def simpson(n):
    h = (b-a)/(2*n)
    nx = new_x_s(n,h)
    sum = 0
    for i in range(2*n):
        sum += (nx[i+1]-nx[i])*(fun(nx[i]) + 4*fun((nx[i+1]+nx[i])/2) + fun(nx[i+1]))/6
    return sum
```

```python
cur = simpson(1)
df.loc[len(df.index)] = [1,0,0,cur]
prev = 0
i = 1
for j in range(8):
    i *= 2
    e = abs(cur - prev)
    prev = cur 
    cur = simpson(i)
    if i == 2:
        df.loc[len(df.index)] = [i,abs(cur-prev),0,cur]
    else:
        df.loc[len(df.index)] = [i,abs(cur-prev),-math.log2(abs(cur-prev)/e),cur]
    print(cur,prev,i)
df
```

Обобщенная формула трапеций

```python
def trap(arr,n):
    integral = 0
    for i in range(0,n):
        integral += (arr[i+1]-arr[i])*(fun(arr[i+1])+fun(arr[i]))
    return integral*0.5
```

```python
cur = trap(new_x((b-a)/1,1),1)
prev = 0
i = 1
while abs(cur-prev) > eps:
    i *= 2
    e = abs(cur - prev)
    prev = cur 
    cur = trap(new_x((b-a)/i,i),i)
    if i == 2:
        df1.loc[len(df1.index)] = [i,abs(cur-prev),0,cur]
    else:
        df1.loc[len(df1.index)] = [i,abs(cur-prev),-math.log2(abs(cur-prev)/e),cur]
print(cur,prev,i)
df1
```

Дифференцирование

```python
def fun_e(x):
    return math.exp(x)
```

```python
h = 1
while d_fun_e(1,h) != 0:
    h /= 2
    print(math.e-d_fun_e(1,h), h)
```

Видно, что при устремлении h к нулю ошибка сначала падает, но начиная с некторого h опять начинает расти 

Добавим шум в данные

```python
def sd_fun_e(x,h):
    return (fun_e(x+h+0.0000001) - fun_e(x+0.0000002))/h
```

```python
h = 1
while d_fun_e(1,h) != 0:
    h /= 2
    print(math.e - sd_fun_e(1,h), h)
```
