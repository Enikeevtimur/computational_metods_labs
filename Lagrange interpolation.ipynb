```python
def new_x(n):
    arr = []
    for i in range(n+1):
        arr.append((2*i)/n - 1)
    return arr
```

```python
new_x(20)
```

```python
def lagrange(x,k):
    f = 0
    for i in range(k):
        p = 1
        for j in range(k):
            if i != j:
                p *= (x-new_x(k)[j])/(new_x(k)[i]-new_x(k)[j])
        f += p*func(new_x(k)[i])
    return f
```

```python
def func(x):
    return 1/(1+25*(x**2))
    
```

```python
x = np.linspace(-1,1,10**5)
plt.plot(x,1/(1+25*x**2),x,lagrange(x,5))
plt.show()
```
